```python
some_energies_consumption = ["nuclear_consumption","oil_consumption","fossil_fuel_consumption","solar_consumption","hydro_consumption","wind_consumption","gas_consumption"]
consumption_df=df[some_energies_consumption]
consumption_df.loc[(df['country'] == 'Canada') & (df['year'] == 2018)]
```

```python
turkey_2018 = consumption_df.loc[(df['country'] == 'Turkey') & (df['year'] == 2018)]
total_energy_consumption = turkey_2018['nuclear_consumption'] + turkey_2018['oil_consumption'] + turkey_2018['solar_consumption']+turkey_2018["fossil_fuel_consumption"]+turkey_2018["hydro_consumption"]+turkey_2018["wind_consumption"]+turkey_2018["gas_consumption"]
print("TÃ¼rkiye'nin 2018'deki toplam enerji tÃ¼ketimi:", total_energy_consumption.sum())
x=round(total_energy_consumption.sum(),2)
print("yuvarlanmÄ±ÅŸ hali",x)
```

```python
df_2019 = df[df['year'] == 2019]
df_2018 = df[df['year'] == 2018]
toplam_enerji_tuketimi_2018 = df_2018[some_energies_consumption].sum()
toplam_enerji_tuketimi_2019 = df_2019[some_energies_consumption].sum()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].pie(toplam_enerji_tuketimi_2018, labels=toplam_enerji_tuketimi_2018.index, autopct='%1.1f%%', startangle=90)
axs[0].set_title('Energy consumption in 2018')

axs[1].pie(toplam_enerji_tuketimi_2019, labels=toplam_enerji_tuketimi_2019.index, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Energy consumption in 2019')

plt.show()
```
